```python
df = df[['date'] + [col for col in df.columns if col != 'date']]
df.head()
```

```python
df.isnull().sum()
```

```python
df['date'] = pd.to_datetime(df['date'])
df = df.sort_values(['provincia', 'date'])

```

```python
def imputar_grupo(g):
    g = g.set_index('date')
    if 'precipitation_sum' in g.columns:
        g['precipitation_sum'] = g['precipitation_sum'].ffill().bfill()
    for col in g.columns:
        if col not in ['provincia', 'precipitation_sum']:
            g[col] = g[col].interpolate(method='time', limit_direction='both')
    return g.reset_index()
```

```python
df_imputacion_nulos = df.groupby('provincia').apply(imputar_grupo).reset_index(drop=True)
df_imputacion_nulos
```

```python
df_imputacion_nulos.isnull().sum()
```

```python
df_imputacion_nulos['day_of_year'] = df_imputacion_nulos['date'].dt.dayofyear
df_imputacion_nulos
```

```python
df_imputacion_nulos['sin_day_of_year'] = np.sin(2 * np.pi * df_imputacion_nulos['day_of_year'] / 366)
df_imputacion_nulos['cos_day_of_year'] = np.cos(2 * np.pi * df_imputacion_nulos['day_of_year'] / 366)

df_imputacion_nulos
```

```python
df_imputacion_nulos.drop(['day_of_year'], axis=1, inplace=True)
```

```python
df_imputacion_nulos.to_csv('data_transformed_open_meteo.csv', index=False)
```
