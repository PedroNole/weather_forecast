```python
df = df.sort_values(by=['fecha']).reset_index(drop=True)
df
```

```python
df.isnull().sum()
```

```python
vars_interes = ['velmedia', 'velmax', 'tmed', 'tmax', 'tmin']
df['COMPLETO'] = df[vars_interes].notnull().all(axis=1)
```

```python
df_check = (
    df.groupby(['provincia', 'fecha'])['COMPLETO']
      .any()  # Devuelve True si al menos una fila del grupo es True
      .reset_index()
      .rename(columns={'COMPLETO': 'TIENE_COMPLETO'})
)
```

```python
df_dia_incompleto = df_check[df_check['TIENE_COMPLETO'] == False]

print("Días/provincias sin ninguna estación con datos completos en ese día:")
df_dia_incompleto
```

```python
df_2 = df[~df['provincia'].isin(['CEUTA', 'MELILLA'])]
df_2
```

```python
df_2['COMPLETO'] = df_2[vars_interes].notnull().all(axis=1)

df_2_check = (
    df_2.groupby(['provincia', 'fecha'])['COMPLETO']
      .any()  # Devuelve True si al menos una fila del grupo es True
      .reset_index()
      .rename(columns={'COMPLETO': 'TIENE_COMPLETO'})
)

df_2_incomplete_days = df_2_check[df_2_check['TIENE_COMPLETO'] == False]

print("Días/provincias sin ninguna estación con datos completos:")
df_2_incomplete_days
```

```python
df_2.dtypes
```

```python
df_2
```

```python
for var in vars_interes:

    df_2[var] = df_2[var].astype(str)
    df_2[var] = df_2[var].str.replace(',', '.', regex=False)
    df_2[var] = pd.to_numeric(df_2[var], errors='coerce')
    
df_2
```

```python
for var in vars_interes:
    df_2[var] = df_2.groupby(['provincia', 'fecha'])[var].transform(lambda x: x.fillna(x.mean()))
```

```python
df_2
```

```python
df_2.drop('COMPLETO', axis=1, inplace=True)
```

```python
df_2.isnull().sum()
```

```python
df_2.isna().sum()
```

```python
df_groupby_prov_fechas = (
    df_2
    .groupby(['provincia', 'fecha'])
    .agg({
        'velmedia': 'mean',
        'velmax': 'mean',
        'tmed': 'mean',
        'tmax': 'mean',
        'tmin': 'mean'
    })
    .reset_index()
)

df_groupby_prov_fechas
```

```python
df_groupby_prov_fechas['fecha'] = pd.to_datetime(df_groupby_prov_fechas['fecha'], errors='coerce')
df_groupby_prov_fechas['day_of_year'] = df_groupby_prov_fechas['fecha'].dt.dayofyear
```

```python

df_groupby_prov_fechas['sin_day_of_year'] = np.sin(2 * np.pi * df_groupby_prov_fechas['day_of_year'] / 366)
df_groupby_prov_fechas['cos_day_of_year'] = np.cos(2 * np.pi * df_groupby_prov_fechas['day_of_year'] / 366)
```

```python
df_groupby_prov_fechas
```

```python
df_groupby_prov_fechas.drop(['fecha', 'day_of_year'], axis=1, inplace=True)
df_groupby_prov_fechas
```

```python
df_groupby_prov_fechas.to_csv('tem_historico_groupby_sincos.csv', index=False)
```
