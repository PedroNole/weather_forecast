```python
import pandas as pd
df = pd.read_csv('clean_madrid_data.csv')
df
```

```python
df_clean = pd.pivot_table(df, index=['ESTACION_NOMBRE','DATE'], columns='MAGNITUD', values='VALUE', aggfunc='mean')
df_clean
```

```python
df_clean.isnull().sum()
```

```python
df_clean
```

```python
df_clean = df_clean.reset_index()
df_completo = df_clean[['ESTACION_NOMBRE', 'DATE', 'HUMEDAD RELATIVA', 'TEMPERATURA', 'PRECIPITACIÓN']]
df_completo
```

```python
df_completo.dropna(subset=['TEMPERATURA', 'HUMEDAD RELATIVA'], inplace=True)
df_completo['PRECIPITACIÓN'] = df_completo['PRECIPITACIÓN'].fillna(0)
```

```python
df_completo.isnull().sum()
```

```python
df_completo
```

```python
df_completo.to_csv('../../transformed/madrid.csv', index=False)
```
